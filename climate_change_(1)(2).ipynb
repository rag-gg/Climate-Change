```python
import pandas as pd                               # For DataFrame variables and Reading a CSV File for processing
import matplotlib.pyplot as plt                   # For plotting the graphs
from sklearn.linear_model import LinearRegression # For Linear Regression
```

```python
# Reading the CSV file

# 'df' is a variable of type DataFrame from the Pandas module
df = pd.read_csv('./data.csv')
df.head()
```

```python
# 'df_new' is a variable holding the data after removing uneccesary columns.
df_new = df.drop(columns = ['Domain', 'Domain Code', 'Element Code', 'Element', 'Months', 'Year Code', 'Unit','Flag','Flag Description'])
df_new.head()
```

**Q1. Whats the change in mean surface temperature for 5 countries(USA,India,Singapore,UAE,Canada) in the last 60 years?**

```python
'''
    Plotting graphs for change in mean surface temperature for the following 5 countries.
'''
# USA
USA_df = df_new[df_new['Area Code (FAO)'] ==  231]
title_value = USA_df['Area'].iat[0]
fig = USA_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')

# India
India_df = df_new[df_new['Area Code (FAO)'] ==  100]
title_value = India_df['Area'].iat[0]
fig = India_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')

# Singapore
Singapore_df = df_new[df_new['Area Code (FAO)'] ==  200]
title_value = Singapore_df['Area'].iat[0]
fig = Singapore_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')

# United Arab Emirates
UAE_df = df_new[df_new['Area Code (FAO)'] ==  225]
title_value = UAE_df['Area'].iat[0]
fig = UAE_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')

# Canada
Canada_df = df_new[df_new['Area Code (FAO)'] ==  33]
title_value = Canada_df['Area'].iat[0]
fig = Canada_df.plot.scatter(x='Year',y='Value', title=title_value)
fig.set_ylabel('Mean surface temperature change(°C)')
```

```python
'''
    Removing all the empty records before performing Linear Regression
'''
Singapore_df = Singapore_df.dropna()
USA_df = USA_df.dropna()
Canada_df = Canada_df.dropna()
UAE_df = UAE_df.dropna()
India_df = India_df.dropna()
```

**Q2. What is the change in temperature from today to 2050 in the countries?**

```python
'''
    Performing Linear Regression to predict the future Mean Surface Temperature Change for countries
'''
# USA
X = USA_df.iloc[:, 3].values.reshape(-1, 1)
Y = USA_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2050, The predicted mean surface temperature change for USA would be {lr.predict([[2050]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('USA')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()


#Singapore
X = Singapore_df.iloc[:, 3].values.reshape(-1, 1)
Y = Singapore_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2050, The predicted mean surface temperature change for Singapore would be {lr.predict([[2050]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('Singapore')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()


#India
X = India_df.iloc[:, 3].values.reshape(-1, 1)
Y = India_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2050, The predicted mean surface temperature change for India would be {lr.predict([[2050]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('India')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()


#UAE
X = UAE_df.iloc[:, 3].values.reshape(-1, 1)
Y = UAE_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2050, The predicted mean surface temperature change for United Arab Emirates would be {lr.predict([[2050]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('UAE')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()


#Canada
X = Canada_df.iloc[:, 3].values.reshape(-1, 1)
Y = Canada_df.iloc[:, 4].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
print(f'In the year 2050, The predicted mean surface temperature change for Canada would be {lr.predict([[2050]])[0][0]}°C')

plt.scatter(X, Y)
plt.title('Canada')
plt.ylabel('Mean surface temperature change(°C)')
plt.xlabel('Year')
plt.ylim(-4,4)
plt.plot(X, Y_pred, color='red')
plt.show()
```
